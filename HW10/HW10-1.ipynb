# Why the energy estimate fails for $$ u(x,t)\equiv 1 $$

Consider the PDE

$$
u_t = \nu u_{xx} + f(x,t), \qquad f \equiv 0.
$$

Let the exact solution be

$$
u(x,t) = 1.
$$

The energy estimate derived earlier was

$$
E(t) \le e^{-\gamma t}E(0)
    + \frac{1}{\gamma}\int_0^t e^{\gamma(s-t)}F(s)\,ds,
\tag{2}
$$

where

$$
E(t)=\int_0^1 u^2(x,t)\,dx, 
\qquad 
F(t)=\int_0^1 f^2(x,t)\,dx.
$$

---

## 1. The estimate requires the Poincaré inequality

The derivation of (2) uses the differential inequality

$$
E'(t) + \gamma E(t)
\le \gamma^{-1}\|f\|_2^2,
\qquad 
\gamma = \frac{2\nu}{C_p^2} > 0.
$$

This comes from the Poincaré inequality:

$$
\int_0^1 u^2\,dx 
\le C_p^2 \int_0^1 (u_x)^2\,dx.
$$

This inequality **holds only** for:

- functions satisfying homogeneous Dirichlet boundary conditions, or  
- functions with zero mean.

---

## 2. Why it fails for $ u \equiv 1 $

For the constant solution $ u(x,t)=1 $:

$$
u_x = 0,
\qquad 
E(t) = \int_0^1 1^2\,dx = 1,
\qquad 
E'(t) = 0.
$$

But (2) predicts:

$$
E(t) \le e^{-\gamma t} E(0) = e^{-\gamma t}.
$$

Since $ \gamma > 0 $, the right-hand side decays to $ 0 $, whereas the true energy satisfies:

$$
E(t) \equiv 1.
$$

This is a contradiction.

---

## 3. Root cause

The failure occurs because the Poincaré inequality **does not hold for constant functions**.

The coercivity estimate used in the derivation,

$$
\int_0^1 (u_x)^2\,dx 
\ge \frac{1}{C_p^2} E(t),
$$

is invalid here:

- Left-hand side = 0  
- Right-hand side = $1/C_p^2 > 0$

Thus, the constant solution violates the assumptions needed to obtain a positive $\gamma$.  
Without $\gamma>0$, the exponential decay estimate (2) is not valid.

---

## 4. Conclusion

The energy estimate (2) does **not** hold for the constant solution $ u \equiv 1 $ because the derivation relies on a Poincaré inequality that requires boundary conditions (or zero-mean conditions) not satisfied by the constant function.

As a result, the coercivity—and therefore the exponential decay—fails.
